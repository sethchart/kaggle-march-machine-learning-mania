```python
import pandas as pd
```

Let's load in the dataframe of tournament wins and team IDs.

```python
seth_df = pd.read_csv('../../data/processed/tournament_outcomes.csv')
```

```python
seth_df.head()
```

```python
seth_df = seth_df.drop(['Unnamed: 0'],axis=1)
seth_df.head()
```

```python
df = pd.read_csv('../../data/processed/data_averages.csv')
```

```python
df.head()
```

Let's merge the low ID team with it's data by season. To do that we'll temporarily rename LowID to TeamID and merge on Season and TeamID. 

```python
LowID = pd.merge(seth_df.rename({'LowID': 'TeamID'}, axis=1), df, how="left", on=['Season','TeamID']).dropna().reset_index(drop=True)
LowID.head()
```

Now let's rename TeamID to LowID in the LowID dataframe and, and rename colunms to reflect that it's LowID data.

```python
LowID.columns = ['Season','LowID','HighID','Win','LowScore','LowFGP','LowFGP3','LowFTP','LowOR','LowDR','LowAst','LowTO','LowStl','LowBlk','LowPF','LowRank','LowSeed']
```

```python
LowID.head()
```

Let's do the same with the High ID teams as well.

```python
HighID = pd.merge(seth_df.rename({'HighID': 'TeamID'}, axis=1), df, how="left", on=['Season','TeamID']).dropna().reset_index(drop=True)
HighID.head()
```

Now let's rename TeamID to HighID in the HighID dataframe and, and rename colunms to reflect that it's HighID data.

```python
HighID.columns = ['Season','LowID','HighID','Win','HighScore','HighFGP','HighFGP3','HighFTP','HighOR','HighDR','HighAst','HighTO','HighStl','HighBlk','HighPF','HighRank','HighSeed']
```

```python
HighID.head()
```

And finally we can merge the HighID and LowID DataFrame on Season, LowID, HighID, and Win since those should be the same in each DataFrame.

```python
model_dataset = pd.merge(LowID, HighID, how='outer', on=['Season','LowID','HighID','Win'])
model_dataset.head()
```

```python
model_dataset.shape
```

Excellent, this is the dataset that we will use in our classifier model where we use all of the features to predict the probability of a Win. So the target is the Win column. Let's save it!

```python
model_dataset.to_csv('../../data/processed/model_dataset.csv')
```
